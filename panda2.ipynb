```python
## CONTAR QTOS PASSAGEIRPS SOBREVOVERAM E QTOS NÃO
sobreviventes = df['Survived'].value_counts()
sobreviventes
```

```python
## CALCULAR TAXA DE SOBREVIVENTES POR SEXO (usar groupby)

sobreviventes_por_sexo = df.groupby('Sex')['Survived'].mean()
```

```python
sobreviventes_por_sexo*100 
```

```python
## QUAL A IDADE MÉDIA DOS SOBREVIVENTES
idade_media_sobreviventes = df.groupby('Survived')['Age'].mean()
```

```python
idade_media_sobreviventes
```

```python
## QUEM PAGOU A TARIFA MAIS CARA?
tarifa_mais_cara = df.loc[df['Fare'] == df['Fare'].max()]
```

```python
tarifa_mais_cara
```

```python
## MÉDIA DE TARIFAS POR CLASSE
tarifa_por_classe = df.groupby('Pclass')['Fare'].mean()
```

```python
tarifa_por_classe
```

```python
sobreviventes_por_classe = df.groupby('Pclass')['Survived'].sum()
sobreviventes_por_classe
```

```python
## CRIAR GRAFICO DE BARRAS MOSTRANDO A SOBREVIVENCIA POR CLASSE
import matplotlib.pyplot as plt

plt.bar(sobreviventes_por_classe.index, sobreviventes_por_classe.values)
plt.xlabel("Classe")
plt.ylabel("Número de sobreviventes")
plt.title("Sobreviventes por classe")
plt.show()
```

```python
## EXERCICIOS

## passageiros que embarcaram em qual porto tiveram maior taxa de sobrevivência
taxa_sobreviventes_por_porto_embarque = df.groupby('Embarked')['Survived'].mean()*100
taxa_sobreviventes_por_porto_embarque
```

```python
## qual a idade média dos passageiros por título (Mr., miss, master, etc.)
def extract_title(name):
    if '.' in name:
        return name.split(',')[1].split('.')[0].strip()

df['Title'] = df['Name'].apply(extract_title)

idade_por_titulo = df.groupby('Title')['Age'].mean().reset_index()
idade_por_titulo
```
