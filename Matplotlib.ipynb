```python
import matplotlib.pyplot as plt
```

```python
### 1) GRÁFICO DE LINHAS: Evolução da temperatura ao longo da semana
```

```python
dias = ['Seg', 'Ter', 'Qua', 'Qui', 'Sex', 'Sab', 'Dom']
temperaturas = [19, 22, 23, 21, 27, 31, 33]
```

```python
plt.figure(figsize=(8,4))
plt.plot(dias, temperaturas, marker='o', linestyle='--', color='purple')
plt.title("Evolução da temperatura ao longo da semana")
plt.xlabel("Dias da semana")
plt.ylabel("Temperatura (°C)")
plt.grid(True, linestyle=":", alpha=0.3)
plt.show()
```

```python
### 2) GRÁFICO DE LINHAS: Intenção de voto
```

```python
meses = ['Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set']
collor = [5, 14, 25, 20, 19, 22]
itamar = [10, 15, 21, 16, 14, 19]
fhc = [20, 22, 25, 28, 32, 36]
```

```python
plt.figure(figsize=(10,6))
plt.plot(meses, collor, marker='o', linestyle='--', color='purple', label='Fernando Collor')
plt.plot(meses, itamar, marker='x', linestyle='-', color='blue', label='Itamar Franco')
plt.plot(meses, fhc, marker='^', linestyle='-.', color='pink', label='Fernando Henrique Cardoso')


plt.title("Intenções de voto para presidente")
plt.xlabel("Mês da Pesquisa")
plt.ylabel("Intenção de Votos (%)")

plt.ylim(0,50)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

```python
### 3) GRÁFICO DE BARRAS: Comparar a quantidade de produtos vendidos em diferentes categorias
```

```python
categorias = ['Eletrônicos', 'Roupas', 'Alimentos', 'Livros']
vendas = [120, 95, 210, 80]
```

```python
plt.figure(figsize=(8,4))

plt.bar(categorias, vendas, color=['purple', 'blue', 'red', 'pink'])
plt.title("Vendas por categorias de produtos")
plt.xlabel("Categoria de produtos")
plt.ylabel("Quantidade vendida")

plt.ylim(8, 240)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
### 4) HISTOGRAMA: Analisar a distribuição de notas de uma turma de 500 alunos
```

```python
import numpy as np
np.random.seed(42)
notas = np.random.normal(70, 10, 500)
```

```python
plt.figure(figsize=(8,5))
plt.hist(notas, bins=50, edgecolor='black', color='purple')
plt.title("Distribuição de notas da turma")
plt.xlabel("Notas")
plt.ylabel("Número de alunos")
plt.axvline(notas.mean(), color='red', linestyle='--', linewidth='2')
plt.legend(['Média das notas'])

plt.grid(True, linestyle=":", alpha=0.5)
plt.ylim(0,40)
plt.show()
```

```python
### 5) GRÁFICO DE PIZZA: Analisar a composição de uma carteira de investimentos
```

```python
ativos = ['Ações', 'Renda Fixa', 'Ouro', 'Criptomoedas']
valores = [45, 30, 15, 10]
```

```python
plt.figure(figsize=(5,5))

explode = (0.1, 0, 0, 0)
plt.pie(valores, labels=ativos, startangle=90, autopct='%1.1f%%', explode=explode)
plt.title('Composição da carteira de investimentos')
plt.axis('equal')
plt.show()
```

```python
### 6) GRÁFICO DE DISPERSÃO: Analisar a relação entre a altura e o peso de um grupo de pessoas
```

```python
np.random.seed(0)
alturas = np.random.normal(170, 10, 50)
pesos = (alturas * 0.5) + np.random.normal(0,5,50)
```

```python
plt.figure(figsize=(8,4))
plt.scatter(alturas, pesos, color='purple', s=20, alpha=0.5)
plt.title("Relação entre altura e peso")
plt.xlabel("Altura (cm)")
plt.ylabel("Peso (kg)")
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

```python
## 01) Comparar contagem de sobreviventes e não sobreviventes divididos por gênero

import pandas as pd
df = pd.read_csv("titanic.csv")
```

```python
sobreviventes = df['Survived'].value_counts()
sobreviventes
```

```python
sobreviventes_por_sexo = df.groupby('Sex')['Survived'].sum()
sobreviventes_por_sexo
```

```python
contagem_sobreviventes_sexo = df.groupby(['Sex', 'Survived']).size().unstack()
```

```python
contagem_sobreviventes_sexo.plot(kind='bar', figsize=(10, 6), rot=0)

plt.title('Contagem de Sobreviventes e Não Sobreviventes por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Número de Sobreviventes')
plt.xticks(ticks=[0, 1], labels=['Feminino', 'Masculino'])
plt.legend(title='Sobrevivência', labels=['Não Sobreviventes', 'Sobreviventes'])

plt.show()
```
